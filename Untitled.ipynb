```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import hashlib

# --------------------------
# Helper functions
# --------------------------
def name_seed(name: str) -> int:
    """Hash the name into an integer seed."""
    h = hashlib.sha256(name.encode()).hexdigest()
    return int(h[:8], 16)

def generate_sierpinski(iterations, seed):
    np.random.seed(seed)
    vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])
    point = np.random.rand(2)
    points = []

    for _ in range(iterations):
        vertex = vertices[np.random.randint(0, 3)]
        point = (point + vertex) / 2
        points.append(point)

    return np.array(points)

# --------------------------
# Get user input
# --------------------------
name = input("Enter your name: ").strip() or "Ariel"
fav_raw = input("Enter your favorite number (default: 13): ").strip() or "13"

try:
    favorite = int(fav_raw)
except ValueError:
    print("Invalid number. Using default (13).")
    favorite = 13

# --------------------------
# Generate and plot fractal
# --------------------------
seed = name_seed(name)
iterations = 10000 + (favorite % 50) * 1000  # Range: 10k to 60k points
points = generate_sierpinski(iterations, seed)

plt.figure(figsize=(8, 8))
plt.scatter(points[:, 0], points[:, 1], s=0.1, color='blue')
plt.axis('off')
plt.title(f"{name}'s Sierpinski Triangle ({iterations} points)")
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import hashlib
import random

# --------------------------
# Helper functions
# --------------------------
def get_seed_from_name(name: str) -> int:
    """Hash name to get deterministic seed."""
    h = hashlib.sha256(name.encode()).hexdigest()
    return int(h[:8], 16)

def draw_branch(x, y, angle, length, depth, max_depth, ax, twist=0.0, chaos=0.0):
    if depth > max_depth or length < 0.5:
        return
    
    # Calculate new branch endpoint
    rad = np.radians(angle)
    x2 = x + length * np.cos(rad)
    y2 = y + length * np.sin(rad)
    
    # Color and linewidth vary with depth
    color = plt.cm.magma(depth / max_depth)
    linewidth = 1.5 * (1 - depth / max_depth)
    
    ax.plot([x, x2], [y, y2], color=color, linewidth=linewidth)
    
    # Recursive branching
    next_length = length * (0.7 + random.uniform(-chaos, chaos))
    delta = 25 + twist * np.sin(depth * 1.2)
    
    draw_branch(x2, y2, angle - delta, next_length, depth + 1, max_depth, ax, twist, chaos)
    draw_branch(x2, y2, angle + delta, next_length, depth + 1, max_depth, ax, twist, chaos)

# --------------------------
# Get user input
# --------------------------
name = input("Enter your name: ").strip() or "Ariel"
fav_raw = input("Enter your favorite number (default: 13): ").strip() or "13"

try:
    favorite = int(fav_raw)
except ValueError:
    print("Invalid number. Using default (13).")
    favorite = 13

# Seed randomness
seed = get_seed_from_name(name)
random.seed(seed)

# Parameters based on input
depth = 5 + (favorite % 5)           # recursion depth 59
twist = (seed % 100) / 50 - 1        # twist between -1 and +1
chaos = (favorite % 7) * 0.05        # small randomness in length

# --------------------------
# Draw the glitched fractal tree
# --------------------------
fig, ax = plt.subplots(figsize=(8, 10))
ax.set_facecolor("black")
ax.axis('off')

# Start from center bottom
draw_branch(0, -1, 90, 1.5, 0, depth, ax, twist=twist, chaos=chaos)

plt.title(f"{name}'s Glitched Fractal Tree (depth={depth}, twist={twist:.2f})", color='white')
plt.show()

```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import hashlib

# ------------------------
# Personalization helpers
# ------------------------
def hash_to_float(h, start=0, end=8, scale=1.0, shift=0.0):
    return int(h[start:end], 16) / 0xFFFFFFFF * scale + shift

def get_fractal_params(name, fav_number):
    h = hashlib.sha256(name.encode()).hexdigest()
    angle_shift = hash_to_float(h, 0, 8, scale=2*np.pi)
    color_offset = hash_to_float(h, 8, 16, scale=1)
    trap_strength = 0.005 + (fav_number % 10) * 0.001
    max_iter = 100 + (fav_number % 20) * 25
    return angle_shift, color_offset, trap_strength, max_iter

# ------------------------
# Orbit Trap Fractal Logic
# ------------------------
def orbit_trap_fractal(xmin, xmax, ymin, ymax, width, height, angle_shift, trap_strength, max_iter, color_offset):
    x = np.linspace(xmin, xmax, width)
    y = np.linspace(ymin, ymax, height)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    C = Z.copy()
    
    trap = np.zeros(Z.shape)
    orbit = np.zeros(Z.shape)
    
    for i in range(max_iter):
        Z = Z ** 2 + C
        angle = np.angle(Z + 1e-9) + angle_shift
        trap = np.minimum(trap + np.abs(np.sin(angle)), 1)
        orbit += np.exp(-trap_strength * np.abs(Z))
    
    norm = (np.sin(orbit + color_offset * np.pi) + 1) / 2
    return norm

# ------------------------
# Get Input
# ------------------------
name = input("Enter your name: ").strip() or "Ariel"
fav_raw = input("Enter your favorite number (default 13): ").strip() or "13"
try:
    favorite = int(fav_raw)
except ValueError:
    favorite = 13

# ------------------------
# Generate Fractal
# ------------------------
angle_shift, color_offset, trap_strength, max_iter = get_fractal_params(name, favorite)
fractal = orbit_trap_fractal(-2, 2, -2, 2, 800, 800, angle_shift, trap_strength, max_iter, color_offset)

plt.figure(figsize=(8, 8))
plt.imshow(fractal, cmap='nipy_spectral', extent=[-2, 2, -2, 2])
plt.axis('off')
plt.title(f"{name}'s Orbit Trap Fractal", fontsize=14)
plt.show()
```
